```python
import pygmt
import numpy as np
import pandas as pd
```

```python
earthquakes = pd.read_csv('/Users/sydneydybing/StrainProject/2024/M6_catalog_2004-2024.csv', dtype = str)
bsm_mda = pd.read_csv('/Users/sydneydybing/StrainProject/2024/bsm_metadata.csv')
```

```python
cape_mend = earthquakes.iloc[[21]]
cape_mend
```

```python
sel_stas = ['B027', 'B028', 'B031', 'B032', 'B035', 'B036', 'B039', 'B040', 'B045', 'B933', 'B934', 'B935']
stas_sub = bsm_mda.query('BNUM in @sel_stas').reset_index(drop = True)
stas_sub
```

```python
# Cape Mend 1

pygmt.config(COLOR_BACKGROUND = 'white')

fig = pygmt.Figure()

region=[-127,-120,39,45]

fig.coast(region=region, projection="N12c", land="wheat3", water="lightcyan2", frame="afg")

# Add plate boundaries

fig.plot(data = '/Users/sydneydybing/StrainProject/2024/PB2002_boundaries.gmt', pen = '1p,darkred', style = 'f0.25/3p', fill = 'darkred')

# Plot earthquakes and stations

fig.plot(x = cape_mend.longitude.astype(float), y = cape_mend.latitude.astype(float), style="+0.7c", fill='white', pen="3p,black", transparency = 20)
fig.plot(x = stas_sub.LONG, y = stas_sub.LAT, style = "t0.7c", fill = "yellow", pen = "black")
fig.text(x = stas_sub.LONG+0.55, y = stas_sub.LAT, text = stas_sub.BNUM, font = '12p,black') 

# Show or save 

# fig.show();

fig.savefig('/Users/sydneydybing/StrainProject/2024/figures/radiation_pattern_overlays/2014-03-10_M6.8_basemap.pdf');

```
