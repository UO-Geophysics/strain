```python
from obspy.core import Stream, read, UTCDateTime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os 
import os.path
```

```python
path_to_files = '/Users/sydneydybing/StrainProject/2024/'
small_test = False

# Events
event_file = 'M6_catalog_2004-2024.csv'
earthquakes = pd.read_csv(path_to_files + event_file, dtype = str)
rows_for_extra_times = [8,9,14,16] # Events that need more identifying info because of duplicate dates
if small_test:
    earthquakes = earthquakes.iloc[[0,1,2]]
origin_times = earthquakes.time.values
mags = earthquakes.mag.values

# Stations
bsm_mda = pd.read_csv(path_to_files + 'bsm_metadata.csv')
stas = bsm_mda.loc[bsm_mda['LONG'] < -110] # North American stations only
# if small_test:
#     stas = stas.iloc[[0,1,2,3,4,5]]
```

```python
theo_p_arrs = np.load(path_to_files + 'theoretical_p_arrivals.npy')
if small_test:
    theo_p_arrs = theo_p_arrs[0:3]
theo_p_arrs.shape
```

```python
UTC_ot_list = []
events_list = []
stas_list = []
diffs_list = []
theo_list = []
manual_list = []

for idx in range(len(earthquakes)):
    
    no_data = False
    
    event_theo_p_arrs = theo_p_arrs[idx]
    
    ot = origin_times[idx]
    mag = mags[idx]
    
    UTC_ot = UTCDateTime(ot)
    
    if idx in rows_for_extra_times:
        cut_ot = ot[:13]
    else:
        cut_ot = ot[:10]
    
    # print('------------------------')
    # print('Event ' + str(cut_ot) + 'M' + str(mag) + ' ' + str(idx+1) + '/' + str(len(earthquakes)))
    
    try:
    
        pick_path = path_to_files + 'seismic_data/processed/' + str(cut_ot) + '_M' + str(mag) + '/picks'
        pick_file = np.genfromtxt(pick_path, dtype = str)
        pick_file_stainfo = pick_file[:,3]

        pick_file_stas = []

        for idx2 in range(len(pick_file_stainfo)):

            pick_file_sta = pick_file_stainfo[idx2].split('.')[1]
            pick_file_stas.append(pick_file_sta)

        pick_file_stas = np.array(pick_file_stas)
        
        # print('')
        # print('Stas in pick file:')
        # print(pick_file_stas)
        # print('')
    
    except: # No pick file
        # print('No data for ' + str(cut_ot) + '_M' + str(mag))
        no_data = True
        continue
    
    for idx3 in range(len(stas)):
  
        sta_theo_p_arr = event_theo_p_arrs[idx3]
        
        try:
            
            sta = stas.BNUM.values[idx3]
            i = np.where(pick_file_stas == sta)[0]
            pick_date = pick_file[i,0][0]
            pick_time = pick_file[i,1][0]
            pick_datetime = UTCDateTime(str(pick_date) + 'T' + str(pick_time) + 'Z')
            
            theo_datetime = UTC_ot + sta_theo_p_arr
            diff = pick_datetime - theo_datetime
            
            # print(sta)
            # print('Manual pick:')
            # print(pick_datetime)
            # print('Theoretical pick:')
            # print(theo_datetime)
            # print('Difference:')
            # print(diff)
            # print('')
            
            UTC_ot_list.append(UTC_ot)
            events_list.append(str(cut_ot) + '_M' + str(mag))
            stas_list.append(sta)
            diffs_list.append(np.abs(diff))
            theo_list.append(theo_datetime)
            manual_list.append(pick_datetime)
        
        except:
            pass
            # print('Error')
            
            
    
    
```

```python
plt.hist(diffs_list, bins = 1000);
plt.xlim(0,10);
```

```python
pick_diff_array = np.column_stack((np.array(UTC_ot_list), np.array(events_list)))
pick_diff_array = np.column_stack((pick_diff_array, np.array(stas_list)))
pick_diff_array = np.column_stack((pick_diff_array, np.array(manual_list)))
pick_diff_array = np.column_stack((pick_diff_array, np.array(theo_list))) 
pick_diff_array = np.column_stack((pick_diff_array, np.array(diffs_list)))
```

```python
for idx in k:
    
    utc_ot = pick_diff_array[idx,0]
    event = pick_diff_array[idx,1]
    sta = pick_diff_array[idx,2]
    manual_pick = pick_diff_array[idx,3]
    theo_pick = pick_diff_array[idx,4]
    
    # print(event)
    # print(sta)
    
    # Load data 
    
    try:
        rms_strain = read(path_to_files + 'strain_data/rms/' + str(event) + '/' + str(sta) + '.mseed')
        peak_strain = read(path_to_files + 'strain_data/peak_post_p/' + str(event) + '/' + str(sta) +'.mseed')
        trimmed_seismic = read(path_to_files + 'seismic_data/processed/' + str(event) + '/' + str(sta) + '/' + 'EHZ.mseed')

        rms_strain_times = rms_strain[0].times()
        rms_strain_data = rms_strain[0].data

        peak_strain_times = peak_strain[0].times()
        peak_strain_data = peak_strain[0].data

        seismic_times = trimmed_seismic[0].times()
        seismic_data = trimmed_seismic[0].data

        manual_time = manual_pick - utc_ot
        theo_time = theo_pick - utc_ot

        fig = plt.figure(dpi = 100)
        ax = fig.add_subplot(111)

        ax.set_title(str(event) + ' ' + str(sta))

        ax.plot(rms_strain_times, rms_strain_data, color = 'lightskyblue', label = 'RMS strain')
        # ax.plot(peak_strain_times, peak_strain_data)
        ax.semilogy()
        
        ax2 = ax.twinx()
        ax2.plot(seismic_times, seismic_data, color = 'lightgreen', label = 'Seismic (EHZ)')
        
        ax.set_xlim(0,120)
        ax2.axvline(manual_time, color = 'black', linestyle = '--', label = 'Manual pick time')
        ax2.axvline(theo_time, color = 'black', label = 'TauPy P-arrival time')
        ax2.legend()
        ax.legend()
        
        plt.show()

    except:
        pass
```
