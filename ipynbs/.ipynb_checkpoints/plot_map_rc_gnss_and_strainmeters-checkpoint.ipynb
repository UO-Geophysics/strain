```python
import numpy as np
# import matplotlib.pyplot as plt
# from obspy.core import read, Stream, UTCDateTime
import pygmt
import pandas as pd
```

```python
path_to_files = '/Users/sydneydybing/StrainProject/2024/'
bsm_mda = pd.read_csv(path_to_files + 'bsm_metadata.csv')
# bsm_mda
```

```python
gnss_stas = pd.read_csv(path_to_files + 'gnss_data/Ridgecrest_GNSS_2019/station_locations.txt', 
                       delimiter = '\s+', skiprows = 1, names = ['sta', 'lon', 'lat', 'elev'])
gnss_stas
```

```python
# US stations

fig = pygmt.Figure()
# fig.coast(region=[-150,210,-90,90], projection = "N12c", land = "wheat4", water = "lightcyan2", frame = "afg")
fig.coast(region=[-119,-116,34.5,36.5], projection = "N12c", land = "lightgray", water = "lightcyan2", frame = "afg")
fig.plot(x = bsm_mda.LONG, y = bsm_mda.LAT, style = "t0.25c", fill = "#F14964", pen = "black")
fig.plot(x = gnss_stas.lon, y = gnss_stas.lat, style = "c0.25c", fill = "#99D9D9", pen = "black")
fig.text(text = bsm_mda.BNUM, x = bsm_mda.LONG+0.7, y = bsm_mda.LAT, font="22p,Helvetica-Bold,#F14964")
fig.text(text = gnss_stas.sta, x = gnss_stas.lon, y = gnss_stas.lat-0.1)
fig.show();
# fig.savefig("/home/sdybing/mlaapde/figures/MLAAPDE_stations_map.png");
```
