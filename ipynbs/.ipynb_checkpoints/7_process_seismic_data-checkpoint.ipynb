```python
client = Client('IRIS')

selected_catalog = np.load('/Users/sydneydybing/StrainProject/2024/NA_evt-sta_dist_array_less500km_withdata.npy')
earthquakes = selected_catalog[:,0]
origin_times = selected_catalog[:,1]
stas = selected_catalog[:,5]

g = np.where(earthquakes == '2023-04-13_M6')

earthquakes_test = earthquakes[g]
origin_times_test = origin_times[g]
stas_test = stas[g]

# Array order:

# 0. Earthquake identifier
# 1. Earthquake origin time
# 2. Earthquake latitude
# 3. Earthquake longitude
# 4. Earthquake magnitude
# 5. Station name
# 6. Station latitude
# 7. Station longitude
# 8. Hypocentral distance
```

```python
i = np.where((earthquakes == '2019-12-23T20_M6') | (earthquakes == '2019-12-23T19_M6') | (earthquakes == '2018-10-22T06_M6.5') | (earthquakes == '2018-10-22T05_M6.5'))[0]

fix_dups = True
if fix_dups:
    earthquakes = earthquakes[i]
    origin_times = origin_times[i]
    stas = stas[i]
```

```python
for idx in range(len(earthquakes)):

    print('Row ' + str(idx+1) + '/' + str(len(earthquakes)))

    eq_id = earthquakes[idx] 
    sta = stas[idx]

    ot = origin_times[idx]
    stime = UTCDateTime(ot)
    etime = stime + 5.*60.
    
    net = 'PB'
    loc = '*'
    chan = 'EHZ'

    try:
        
        path_to_files = '/Users/sydneydybing/StrainProject/2024/seismic_data/raw/' + str(eq_id) + '/' + str(sta) + '/'
#         print(path_to_files + chan + '.mseed')
        data = read(path_to_files + chan + '.mseed')
#         data.plot()

        data_array = data[0].data
        times = data[0].times()

        # Normalizing unfiltered data

        norm_value = data_array[0]
        data_length = data_array[0].shape
        normalize = np.full(data_length, norm_value)

        data_norm = np.subtract(data_array, normalize)

        st_norm = data.copy()
        st_norm[0].data = data_norm
    #     st_norm.plot()

        eq_sta_save_dir = '/Users/sydneydybing/StrainProject/2024/seismic_data/processed/' + str(eq_id) + '/' + str(sta) + '/'
    #         print(eq_sta_save_dir)
        if os.path.isdir(eq_sta_save_dir):
            pass
        else:
            os.makedirs(eq_sta_save_dir)

        st_norm.write(eq_sta_save_dir + chan + '.mseed', format = 'MSEED')

    except:
        pass
#                 print(idx, sta, chan, "not found")
```
