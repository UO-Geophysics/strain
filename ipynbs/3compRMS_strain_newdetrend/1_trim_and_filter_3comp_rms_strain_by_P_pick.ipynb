```python
from obspy.core import Stream, read, UTCDateTime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os 
import os.path
from mudpy import forward
```

```python
path_to_files = '/Users/sydneydybing/StrainProject/2024/'

earthquakes = pd.read_csv(path_to_files + 'M6_catalog_2004-2024.csv', dtype = str)

small_test = False
if small_test:
    earthquakes = earthquakes.iloc[[8]]

weird_events_only = False
if weird_events_only:
    earthquakes = earthquakes.iloc[[4,15]]
    
redo_picks = False
if redo_picks:
    earthquakes = earthquakes.iloc[[3,5,6,8,10,17,19,20,21,24,25,26,27,29]]

bsm_mda = pd.read_csv(path_to_files + 'bsm_metadata.csv')
NA_only = bsm_mda.loc[bsm_mda['LONG'] < -110]
stas = NA_only['BNUM'].values

test = False

if test:
    # stas = ['B027', 'B028', 'B030']
    earthquakes = earthquakes.iloc[[1]]

origin_times = earthquakes.time.values
mags = earthquakes.mag.values    

eqs_test = earthquakes[4:5]
ots_test = origin_times[4:5]
mags_test = mags[4:5]
rows_for_extra_times = [8,9,14,16]
```

```python
for idx in range(len(earthquakes)):
    
    no_data = False
    ot = origin_times[idx]
    print('Earthquake ' + str(idx+1) + '/' + str(len(mags)))
    
    if idx in rows_for_extra_times:
        cut_ot = ot[:13]
    else:
        cut_ot = ot[:10]
        
    print(cut_ot)
        
    try:
    
        pick_path = path_to_files + 'seismic_data/processed/' + str(cut_ot) + '_M' + str(mags[idx]) + '/picks'
        pick_file = np.genfromtxt(pick_path, dtype = str)
        pick_file_stainfo = pick_file[:,3]

        pick_file_stas = []

        for idx2 in range(len(pick_file_stainfo)):

            pick_file_sta = pick_file_stainfo[idx2].split('.')[1]
            pick_file_stas.append(pick_file_sta)

        pick_file_stas = np.array(pick_file_stas)
        # print(pick_file_stas)
    
    except:
        print('No data for ' + str(cut_ot) + '_M' + str(mags[idx]))
        no_data = True
        pass
    
    for sta in stas:
        # print(sta)
        
        try:
        
            i = np.where(pick_file_stas == sta)[0]
            pick_date = pick_file[i,0][0]
            pick_time = pick_file[i,1][0]
            pick_datetime = UTCDateTime(str(pick_date) + 'T' + str(pick_time) + 'Z')
        
        except:
            continue

        # print(sta)    
        starttime = pick_datetime - 10 # 10 seconds before pick time
        endtime = starttime + 120 # full trace will be 2 minutes long

        BS1 = read(path_to_files + 'strain_data/processed/' + str(cut_ot) + '_M' + str(mags[idx]) + '/' + str(sta) + '/BS1.mseed')
        BS2 = read(path_to_files + 'strain_data/processed/' + str(cut_ot) + '_M' + str(mags[idx]) + '/' + str(sta) + '/BS2.mseed')
        BS3 = read(path_to_files + 'strain_data/processed/' + str(cut_ot) + '_M' + str(mags[idx]) + '/' + str(sta) + '/BS3.mseed')

        # High-pass filter
        corner_period = 1000 # seconds
        corner_freq = 1/corner_period

        filterBS1 = BS1.copy()
        filterBS2 = BS2.copy()
        filterBS3 = BS3.copy()

        filterBS1[0].data = forward.highpass(BS1[0].data, fcorner = corner_freq, fsample = BS1[0].stats.sampling_rate, order = 4)
        filterBS2[0].data = forward.highpass(BS2[0].data, fcorner = corner_freq, fsample = BS2[0].stats.sampling_rate, order = 4)
        filterBS3[0].data = forward.highpass(BS3[0].data, fcorner = corner_freq, fsample = BS3[0].stats.sampling_rate, order = 4)

        # Detrend
        # Figure out which sample has the UTCDateTime P-wave arrival
        utc_times = filterBS1[0].times(type = 'utcdatetime')
        diffs = []
        for time in utc_times:
            diff = np.abs(time - pick_datetime)
            diffs.append(diff)
        l = np.where(np.array(diffs) == np.min(np.array(diffs)))[0]
        utcdt_arrival = utc_times[l][0]
        # l = np.where(utc_times == pick_datetime)[0]
        # print('Actual pick:')
        # print(pick_datetime)
        # print('Closest time in data:')
        # print(utcdt_arrival)
        data_p_arr_idx = np.where(utc_times == utcdt_arrival)[0][0]
        # print(data_p_arr_idx)

        filterBS1[0].data = filterBS1[0].data - np.mean(filterBS1[0].data[data_p_arr_idx-2:data_p_arr_idx+2])
        filterBS2[0].data = filterBS2[0].data - np.mean(filterBS2[0].data[data_p_arr_idx-2:data_p_arr_idx+2])
        filterBS3[0].data = filterBS3[0].data - np.mean(filterBS3[0].data[data_p_arr_idx-2:data_p_arr_idx+2])

        RMS_strain = np.sqrt(((filterBS1[0].data)**2 + (filterBS2[0].data)**2 + (filterBS3[0].data)**2)/3)         

        times = filterBS1[0].times()

        RMS_st = filterBS1.copy()
        RMS_st[0].stats.channel = 'BSR'
        RMS_st[0].data = RMS_strain

        # plt.plot(RMS_st[0].data)
        # plt.show()

        # RMS_st.plot()
        # print(RMS_st[0].stats)

        # Create folder for the event
        eq_sta_save_dir = path_to_files + 'strain_data/3comp_rms_newdetrend/' + str(cut_ot) + '_M' + str(mags[idx]) + '/'
#                 print(eq_sta_save_dir)
        if os.path.isdir(eq_sta_save_dir):
            pass
        else:
            os.makedirs(eq_sta_save_dir)

        RMS_st.write(eq_sta_save_dir + sta + '.mseed', format = 'MSEED')

        # Now trim the data 

        RMS = read(path_to_files + 'strain_data/3comp_rms_newdetrend/' + str(cut_ot) + '_M' + str(mags[idx]) + '/' + str(sta) + '.mseed')
        # RMS.plot()

        RMS_trim = RMS.trim(starttime = starttime, endtime = endtime, pad = True, fill_value = 0)
        # RMS_trim.plot()

        # Extracting just the RMS data, now trimmed properly, and saving it to a 
        # new miniSEED file.

        RMS_times_trim = RMS_trim[0].times()
        RMS_data_trim = RMS_trim[0].data
        RMS_trim[0].stats.channel = 'tBSR'

        # plt.plot(RMS_trim[0].data)
        # plt.show()

        # Plot stations, but avoid plotting the weird stations for the two problem earthquakes

        if idx == 4 and sta == 'B072' or sta == 'B076':
            print('Skipping ' + str(cut_ot) + ' ' + str(sta))

        elif idx == 15 and sta == 'B928':
            print('Skipping ' + str(cut_ot) + ' ' + str(sta))

        else:
            plt.title(str(cut_ot) + '_M' + str(mags[idx]))
            plt.plot(RMS_times_trim, RMS_data_trim, label = sta)

        # Create folder for the event
        eq_sta_save_dir = path_to_files + 'strain_data/3comp_trimmed_rms_newdetrend/' + str(cut_ot) + '_M' + str(mags[idx]) + '/'
#                 print(eq_sta_save_dir)
        if os.path.isdir(eq_sta_save_dir):
            pass
        else:
            os.makedirs(eq_sta_save_dir)

        RMS_trim.write(eq_sta_save_dir + sta + '.mseed', format = 'MSEED')

        # except:
        #     pass
    
    # Plot after each earthquake
    if no_data: 
        pass
    else:
        plt.legend(ncol = 2)
        plt.show()  
```
