```python
from obspy.core import Stream, read
import obspy.signal
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os 
import os.path
from mudpy import forward
```

```python
path_to_files = '/Users/sydneydybing/StrainProject/2024/'

earthquakes = pd.read_csv(path_to_files + 'M6_catalog_2004-2024.csv', dtype = str)

weird_events_only = False
if weird_events_only:
    earthquakes = earthquakes.iloc[[4,15]]

small_test = False
if small_test:
    earthquakes = earthquakes.iloc[[11]]
    
redo_picks = False
if redo_picks:
    earthquakes = earthquakes.iloc[[3,5,6,8,10,17,19,20,21,24,25,26,27,29]]

origin_times = earthquakes.time.values
mags = earthquakes.mag.values

bsm_mda = pd.read_csv(path_to_files + 'bsm_metadata.csv')
NA_only = bsm_mda.loc[bsm_mda['LONG'] < -110]
stas = NA_only['BNUM'].values
# stas = ['B072', 'B079', 'B082', 'B087', 'B916', 'B917', 'B918', 'B921']

eqs_test = earthquakes[4:5]
ots_test = origin_times[4:5]
mags_test = mags[4:5]
rows_for_extra_times = [8,9,14,16]
```

```python
for idx in range(len(earthquakes)):
    
    no_data = False
    
    if idx >= 31: # No data available for 4 earliest earthquakes
        no_data = True
        
    if no_data == False:
        plt.figure(figsize = (10,6), dpi = 300, facecolor = 'white', num = 1, clear = True)
        
    ot = origin_times[idx]
    print('Earthquake ' + str(idx+1) + '/' + str(len(mags)))
    
    if idx in rows_for_extra_times:
        cut_ot = ot[:13]
    else:
        cut_ot = ot[:10]
        
    for sta in stas:
        
        try:
            
            # Strain data

            tRMS = read(path_to_files + 'strain_data/trimmed/' + str(cut_ot) + '_M' + str(mags[idx]) + '/' + str(sta) + '.mseed')

            filterRMS = tRMS.copy()

            # plt.plot(tRMS[0].times(), tRMS[0].data, label = 'RMS data')
            # print(tRMS[0].data[0])

            # 1000s highpass filter
            corner_period = 1000 # seconds
            corner_freq = 1/corner_period

            filterRMS[0].data = forward.highpass(tRMS[0].data, fcorner = corner_freq, fsample = tRMS[0].stats.sampling_rate, order = 4)

            # plt.plot(filterRMS[0].times(), filterRMS[0].data, label = 'Filtered RMS data')
            # print(filterRMS[0].data[0])

            # Detrend

            filterRMS[0].data = filterRMS[0].data - np.mean(filterRMS[0].data[0:50])

            # plt.plot(filterRMS[0].times(), filterRMS[0].data, label = 'Detrended filtered RMS data')
            # plt.axhline(0, linestyle = '--', color = 'gray')
            # plt.legend()
            # print(filterRMS[0].data[0])
            # plt.show()

            # Plot stations, but avoid plotting the weird stations for the two problem earthquakes

            if idx == 4 and sta == 'B072' or sta == 'B076':
                print('Skipping ' + str(cut_ot) + ' ' + str(sta))

            elif idx == 15 and sta == 'B928':
                print('Skipping ' + str(cut_ot) + ' ' + str(sta))

            else:
                plt.title('Earthquake ' + str(cut_ot) + '_M' + str(mags[idx]), fontsize = 18)
                plt.plot(filterRMS[0].times(), filterRMS[0].data, label = sta)
                plt.ylabel('Filtered RMS extensional strain', fontsize = 15)
                plt.xlabel('Time (s)', fontsize = 15)
                plt.xlim(0,120)
                plt.tick_params(labelsize = 13)

            # Create folder for the event
            eq_sta_save_dir = path_to_files + 'strain_data/filtered_trimmed_rms/' + str(cut_ot) + '_M' + str(mags[idx]) + '/'
    #                 print(eq_sta_save_dir)
            if os.path.isdir(eq_sta_save_dir):
                pass
            else:
                os.makedirs(eq_sta_save_dir)

            filterRMS.write(eq_sta_save_dir + sta + '.mseed', format = 'MSEED')

        except:
            pass
#             print(quake + " no station " + sta)
            
    # Plot after each earthquake
    if no_data: 
        pass
    else:
        plt.legend(loc = 'lower right', ncol = 2, fontsize = 12)
        # plt.show()
        plt.savefig(path_to_files + 'figures/filtered_peak_strain_plots/NA_M6_500km/not_rainbow/' + str(cut_ot) + '_M' + str(mags[idx]) + '.png', format = 'PNG')
        plt.close();
```
