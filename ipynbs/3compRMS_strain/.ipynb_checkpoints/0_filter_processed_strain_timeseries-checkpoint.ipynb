```python
from obspy.core import Stream, read
import numpy as np
import matplotlib.pyplot as plt
import os 
import os.path
import pandas as pd
from mudpy import forward
```

```python
path_to_files = '/Users/sydneydybing/StrainProject/2024/'

earthquakes = pd.read_csv(path_to_files + 'M6_catalog_2004-2024.csv', dtype = str)

fix_dups = False
if fix_dups:
    earthquakes = earthquakes.iloc[[8,9,14,16]] # Earthquakes with same day and magnitude
    
weird_events_only = False
if weird_events_only:
    earthquakes = earthquakes.iloc[[4,15]]
    
small_test = False
if small_test:
    earthquakes = earthquakes.iloc[[6]] # An earthquake that has two well behaved stations and two bad

origin_times = earthquakes.time.values
mags = earthquakes.mag.values

bsm_mda = pd.read_csv(path_to_files + 'bsm_metadata.csv')
NA_only = bsm_mda.loc[bsm_mda['LONG'] < -110]
stas = NA_only['BNUM'].values
chans = ['BS1', 'BS2', 'BS3', 'BS4']

if small_test: # An earthquake that has two well behaved stations and two bad
    stas = ['B207', 'B208', 'B944', 'B950']
    
rows_for_extra_times = [8,9,14,16]
```

```python
for idx in range(len(earthquakes)):
    
    ot = origin_times[idx]
    print('Earthquake ' + str(idx+1) + '/' + str(len(mags)))
    
    if idx in rows_for_extra_times:
        cut_ot = ot[:13]
    else:
        cut_ot = ot[:10]
    
    if weird_events_only:
        if idx == 0:
            stas = ['B072', 'B076']

        elif idx == 1:
            stas = ['B928']
    
    for sta in stas:
        
#         print(sta)
                    
        try:

            BS1 = read(path_to_files + 'strain_data/processed/' + str(cut_ot) + '_M' + str(mags[idx]) + '/' + str(sta) + '/BS1.mseed')
            BS2 = read(path_to_files + 'strain_data/processed/' + str(cut_ot) + '_M' + str(mags[idx]) + '/' + str(sta) + '/BS2.mseed')
            BS3 = read(path_to_files + 'strain_data/processed/' + str(cut_ot) + '_M' + str(mags[idx]) + '/' + str(sta) + '/BS3.mseed')
            # BS4 = read(path_to_files + 'strain_data/processed/' + str(cut_ot) + '_M' + str(mags[idx]) + '/' + str(sta) + '/BS4.mseed')

    #             BS1.plot()
    #             BS2.plot()
    #             BS3.plot()
    #             BS4.plot()

            # High-pass filter
            corner_period = 1000 # seconds
            corner_freq = 1/corner_period

            filterBS1 = BS1.copy()
            filterBS2 = BS2.copy()
            filterBS3 = BS3.copy()
            # filterBS4 = BS4.copy()

            filterBS1[0].data = forward.highpass(BS1[0].data, fcorner = corner_freq, fsample = BS1[0].stats.sampling_rate, order = 4)
            filterBS2[0].data = forward.highpass(BS2[0].data, fcorner = corner_freq, fsample = BS2[0].stats.sampling_rate, order = 4)
            filterBS3[0].data = forward.highpass(BS3[0].data, fcorner = corner_freq, fsample = BS3[0].stats.sampling_rate, order = 4)
            # filterBS4[0].data = forward.highpass(BS4[0].data, fcorner = corner_freq, fsample = BS4[0].stats.sampling_rate, order = 4)

            # Detrend
            filterBS1[0].data = filterBS1[0].data - np.mean(filterBS1[0].data[0:50])
            filterBS2[0].data = filterBS2[0].data - np.mean(filterBS2[0].data[0:50])
            filterBS3[0].data = filterBS3[0].data - np.mean(filterBS3[0].data[0:50])
            # filterBS4[0].data = filterBS4[0].data - np.mean(filterBS4[0].data[0:50])

            RMS_strain = np.sqrt(((filterBS1[0].data)**2 + (filterBS2[0].data)**2 + (filterBS3[0].data)**2)/3)         

            times = filterBS1[0].times()

    #             plt.plot(times, BS1[0].data, label = 'BS1')
    #             plt.plot(times, BS2[0].data, label = 'BS2')
    #             plt.plot(times, BS3[0].data, label = 'BS3')
    #             plt.plot(times, BS4[0].data, label = 'BS4')
    #             plt.legend()
    #             plt.show()

    #             timeseries = [0.,300.]
    #             #plt.xlim(timeseries)
    #             timeseries = str(timeseries)

    #            plt.xlabel('Time (s)')
    #            plt.ylabel('RMS Microstrain ($10^{-6}$ m/m)')
    #            plt.title(quake + ' Earthquake at PBO Station ' + sta)
    #            plt.legend(loc = 1)

            RMS_st = filterBS1.copy()
            RMS_st[0].stats.channel = 'BSR'
            RMS_st[0].data = RMS_strain

            # RMS_st.plot()
            # print(RMS_st[0].stats)

            # Create folder for the event
            eq_sta_save_dir = path_to_files + 'strain_data/3comp_rms/' + str(cut_ot) + '_M' + str(mags[idx]) + '/'
    #                 print(eq_sta_save_dir)
            if os.path.isdir(eq_sta_save_dir):
                pass
            else:
                os.makedirs(eq_sta_save_dir)

            RMS_st.write(eq_sta_save_dir + sta + '.mseed', format = 'MSEED')

        except:
            pass
```
